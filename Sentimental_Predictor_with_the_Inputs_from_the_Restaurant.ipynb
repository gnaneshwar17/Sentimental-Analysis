```python
import numpy as np
import pandas as pd
dataset = pd.read_csv('/content/New_Restaurant_Inputs.tsv', delimiter = '\t', quoting = 3)
dataset.head()
```

```python
import re
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
customer=[]

for i in range(0, 100):
  review = re.sub('[^a-zA-Z]', ' ', dataset['Review'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  customer.append(review)
```

```python
# Loading BoW dictionary
from sklearn.feature_extraction.text import CountVectorizer
import pickle
cvFile='c1_BoW_Sentiment_Model.pkl'
cv = pickle.load(open(cvFile, "rb"))
```

```python
X_fresh = cv.transform(customer).toarray()
X_fresh.shape
```

```python
import joblib
classifier = joblib.load('/content/c2_Classifier_Sentiment_Model.pkl')
```

```python
y_pred = classifier.predict(X_fresh)
print(y_pred)
```

```python
dataset['predicted_label'] = y_pred.tolist()
dataset.head()
```

```python
from google.colab import files

# Save the DataFrame to a CSV file
dataset.to_csv("Predicted_Outputs.csv", encoding='UTF-8', index=False)

# Make the file downloadable
files.download("Predicted_Outputs.csv")

```

```python

dataset = pd.read_csv("Predicted_Outputs.csv")
average_value = dataset['predicted_label'].mean()
happy_customers_rate = average_value * 100
print(f"The percentage of Happy Customers is: {happy_customers_rate:.2f}%")
```
